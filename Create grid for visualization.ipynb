```python
import pandas as pd
```

```python
df = pd.read_csv("Data/Police_Incidents_grid.csv")
```

```python
df
```

```python
df_sort_before = df.sort_values(['Latitude_round', 'Longitude_round'])[(df.Day_Delta >= -365) & (df.Day_Delta < 0)]
df_sort_before.head()
```

```python
df_sort_after = df.sort_values(['Latitude_round', 'Longitude_round'])[(df.Day_Delta > 0) & (df.Day_Delta <= 365)]
df_sort_after.head()
```

```python
grid_concat_before = df_sort_before.grid.unique().tolist()
grid_concat_after = df_sort_after.grid.unique().tolist()
```

```python
def get_grid_number(grid_concat_text):
    return grid_concat.index(grid_concat_text)
```

```python
df_sort_before['grid_id'] = df_sort_before.grid.apply(lambda x: get_grid_number(x))
df_grouped_before = df_sort_before.groupby('grid_id')['grid','Latitude_round','Longitude_round']
df_grouped_before = df_grouped_before.agg({'Latitude_round':'max', 'Longitude_round':'max','grid':'count'}).drop(0)
df_grouped_before = df_grouped_before.rename(columns={'grid' : 'count'})
df_grouped_before.head()
```

```python
df_sort_after['grid_id'] = df_sort_after.grid.apply(lambda x: get_grid_number(x))
df_grouped_after = df_sort_after.groupby('grid_id')['grid','Latitude_round','Longitude_round']
df_grouped_after = df_grouped_after.agg({'Latitude_round':'max', 'Longitude_round':'max','grid':'count'}).drop(0)
df_grouped_after = df_grouped_after.rename(columns={'grid' : 'count'})
df_grouped_after.head()
```

```python
df_grouped_before.to_csv("Data/heatmap_before.csv")
df_grouped_after.to_csv("Data/heatmap_after.csv")
```
